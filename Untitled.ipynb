```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("results.csv")
```

```python
df_qos1 = df[df['qos'] == 1]  # Filtering for QoS = 1

# Creating subplots side by side with a shared y-axis
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 6), sharey=True)  # 1 row, 2 columns, shared y-axis

colors = {'Tiny': 'blue', 'Small': 'green', 'Medium': 'red', 'Large': 'purple', 'Huge': 'orange'}
linestyles = {0: 'dashed', 1: 'solid'}

# Grouping by 'size' and 'sync', then plotting the average 'ns' with standard deviation 'curves'
for (size, sync), group in df_qos1.groupby(['size', 'sync']):
    grouped = group.groupby('chunks_size')['ns']
    avg_ns = grouped.mean()
    std_ns = grouped.std()
    linestyle = linestyles[sync]
    label = f"{size} {'sync' if sync else 'async'}"
    ax = ax1 if sync else ax2  # Choose the appropriate axis based on sync value
    
    # Plotting the line with marker
    ax.plot(avg_ns.index, avg_ns, label=label, color=colors[size], linestyle=linestyle, marker='o')
    
    # Plotting the standard deviation 'curve'
    ax.fill_between(avg_ns.index, avg_ns - std_ns, avg_ns + std_ns, alpha=0.2, color=colors[size])

# Set common settings for both subplots
for ax in [ax1, ax2]:
    ax.set_xscale('log')
    ax.set_xticks([32, 64, 128, 256, 512, 1024, 2048, 4096, 8192])
    ax.get_xaxis().set_minor_locator(plt.NullLocator())
    ax.get_xaxis().set_major_formatter(plt.FuncFormatter(lambda val, pos: '{:.0f}'.format(val)))
    ax.set_xlabel('Chunk Size')
    ax.legend()

# Setting the ylabel for the leftmost subplot only
ax1.set_ylabel('ns')

# Setting individual titles
ax1.set_title('Sync Performance Metrics with QoS=1')
ax2.set_title('Async Performance Metrics with QoS=1')

# Adjust layout to prevent overlapping
plt.tight_layout()

# Displaying the plot
plt.show()
```

```python
df_sync0 = df[df['sync'] == 0]
df_sync1 = df[df['sync'] == 1]

# Group by size and aggregate
grouped_sync0 = df_sync0.groupby('size')['ns'].apply(list)
grouped_sync1 = df_sync1.groupby('size')['ns'].apply(list)

# Ensure all size categories are present
sizes = ['Tiny', 'Small', 'Medium', 'Large', 'Huge']
all_data_sync0 = [grouped_sync0.get(size, []) for size in sizes]
all_data_sync1 = [grouped_sync1.get(size, []) for size in sizes]

fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

# Plot for sync = 0
axs[0].violinplot(all_data_sync0, showmeans=False, showmedians=True)
axs[0].set_title('Asyncronous')
axs[0].yaxis.grid(True)
axs[0].set_xticks([y + 1 for y in range(len(all_data_sync0))])
axs[0].set_xticklabels(sizes)
axs[0].set_xlabel('Size')
axs[0].set_ylabel('Client-side Service Time (ns)')
axs[0].set_ylim(bottom=0, top=1e10)  # Set y-axis limit

# Plot for sync = 1
axs[1].violinplot(all_data_sync1, showmeans=False, showmedians=True)
axs[1].set_title('Syncronous')
axs[1].yaxis.grid(True)
axs[1].set_xticks([y + 1 for y in range(len(all_data_sync1))])
axs[1].set_xticklabels(sizes)
axs[1].set_xlabel('Size')
axs[1].set_ylim(bottom=0, top=1e10)  # Set y-axis limit
# axs[1].set_ylabel('ns')  # No need to set ylabel as it's shared with the first plot

plt.show()
```
